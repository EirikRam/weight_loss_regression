```python
import pandas as pd

# Load raw data
df = pd.read_csv('../data/raw/calories.csv')

# Encode gender: male = 1, female = 0
df['Gender'] = df['Gender'].map({'male': 1, 'female': 0})

# Drop ID column
df = df.drop(columns=['User_ID'])

# Check if any nulls appeared
print(df.isnull().sum())

# Define features and label
X = df.drop(columns=['Calories'])   # 7 features now, including encoded Gender
y = df[['Calories']]                # Target

# Save cleaned dataset
cleaned = pd.concat([X, y], axis=1)
cleaned.to_csv('../data/processed/cleaned_data.csv', index=False)

# Preview
cleaned.head()
```
